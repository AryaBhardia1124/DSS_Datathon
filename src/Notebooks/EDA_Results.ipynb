```python
def Checking_missing_values(df, column_check):
  bachelor_degree_cols = [col for col in df.columns if column_check in col]

  missing_bachelor_degrees = df[bachelor_degree_cols].isnull()

  missing_pattern_sum = missing_bachelor_degrees.sum(axis=1)

  unique_missing_sums = missing_pattern_sum.unique()

  is_consistent = all(s == 0 or s == len(bachelor_degree_cols) for s in unique_missing_sums)

  if is_consistent:
      print(f"All {column_check} columns have missing values in the exact same rows (either all are missing or none are missing).")
  else:
      print(f"The {column_check} columns do NOT have missing values in the exact same rows. There are rows where some of these columns are missing while others are not.")

  print(f"\nUnique counts of missing values per row for bachelor degree columns: {unique_missing_sums}")
  print(f"\nNumber of missing rows: {df[bachelor_degree_cols[0]].isna().sum()}")
```
